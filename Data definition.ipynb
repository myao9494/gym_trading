# データの確認


インプットデータを確認する。このプログラムは、FXの対応なので、株への適用は出来高を追加する必要がある。

データの読み込み

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
df=pd.read_csv('data/EURUSD60.csv', header=None, encoding='shift-jis')
pd.set_option('display.max_colwidth', 80)#カラム内の文字数。デフォルトは50だった
pd.set_option('display.max_rows', 500)#行数
pd.options.display.notebook_repr_html = True
df.columns = ['a','b','open','high','low','close','g']  #カラム名を付ける
df
# df=df.drop(['g'],axis=1) #列を削除する
```

読み込んだデータの時間をdatetimeへ変換

```python
from datetime import datetime as dt
for i, v in df.iterrows(): #下記の場合は、X,Yの列名があり、行方向へのループ（列の場合は行名を入れる）
    tstr =v['a'].replace(".","/")+" "+v['b']+":00"   # iは行または列名　v は Series
#     print(tstr)
    tdatetime = dt.strptime(tstr, '%Y/%m/%d %H:%M:%S')
#     print(tdatetime)
    df.loc[i,"a"]=tdatetime
df=df.drop(['b'],axis=1) #列を削除する
```

整理したデータ

```python
df=df.loc[1:100] # 列番号指定
```

```python
df
```

グラフで見える化

```python
plt.style.use('seaborn-darkgrid')# style
plt.figure(figsize=(12, 7))
palette = plt.get_cmap('Set1')# create a color palette
num=0# multiple line plot
for column in df.drop('a', axis=1):
    num+=1
    plt.plot(df['a'], df[column], marker='', color=palette(num), linewidth=1, alpha=0.9, label=column)
plt.legend(loc=2, ncol=2)# Add legend
plt.title('plot', loc='left', fontsize=12, fontweight=0, color='orange')# Add titles
plt.xlabel('date')
plt.ylabel('value')
```
