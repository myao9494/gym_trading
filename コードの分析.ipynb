```python
import pandas as pd
import numpy as np
```

```python
csv_name="data/EURUSD60.csv"
df=pd.read_csv(csv_name, parse_dates=[[0, 1]], header=None,
                             names=['Date', 'Time', 'Open', 'High', 'Low', 'Close', 'Volume'])
df = df[~np.isnan(df['Open'])].set_index('Date_Time')#~は否定演算子
```

### 特徴量エンジニアリング（dfにテクニカル分析結果などを追加する：より特徴が出る指標を追加）
※ここが肝。大量のデータよりも、ここが優れているかが良い学習結果につながるとのこと

```python
def _calculate_return(df):
    '''
    Returnを追加して、計算する。リターンの計算の仕方は不思議な感じ。。。
    '''
    # Compute Returns based on consecutive closed prices
    df['Return'] = df['Close'].pct_change() * 100

    # Normalization of returns
    mean = df['Return'].mean()
    std = df['Return'].std()
    df['Return'] = (df['Return'] - np.array(mean)) / np.array(std)

    return df

def _generate_indicators(data):
    '''
    テクニカル分析結果を追加（ここではATR）
    '''
    _high = data.High.values
    _low = data.Low.values
    _close = data.Close.values

    # Compute the ATR and perform Normalisation
    data['ATR'] = talib.ATR(_high, _low, _close, timeperiod=14)
    data.dropna(inplace=True)
    data['ATR'] = (data['ATR'] - np.mean(data['ATR'])) / np.std(data['ATR'])

    return data

```

#### dfを完成させる
